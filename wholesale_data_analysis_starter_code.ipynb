## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal']=df["qty"] * df["unit_price"]
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

df["shippingprice"] = df["unit_weight"].apply(lambda w: w * 7 if w > 50 else w * 10 )
df
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["totalprice"]=( (df["subtotal"] + df["shippingprice"]) * ( 1 + 0.0925))
df
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df["totalcost"]=((df["unit_cost"] * df["qty"]) + df["shippingprice"]) 
df
```

```python
# Create a column for the profit of each line using line cost and line price
df["profit"] = df["totalprice"] - df["totalcost"]
df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
df_topclient = df.groupby("client_id")["qty"].sum().reset_index()
df_topclient.sort_values("qty",ascending=False).head(5)

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
df_topclientsummary = df.groupby("client_id").agg(TotalUnitsPurchased = ("qty","sum"),TotalShippingPrice = ("shippingprice","sum"),TotalRevenue=("totalprice","sum"),TotalProfit=("profit","sum"))
df_topclientsummary = df_topclientsummary.sort_values("TotalUnitsPurchased",ascending=False).head(5)
df_topclientsummary

```

```python
# Format the data and rename the columns to names suitable for presentation.
from decimal import Decimal 

# Define the money columns. 
df_topclientsummary["TotalShippingPrice"] = df_topclientsummary["TotalShippingPrice"].apply(lambda x: Decimal(x)).map('${:,.2f}'.format)
#df_topclientsummary["TotalRevenue"] = df_topclientsummary["TotalRevenue"].apply(lambda x: Decimal(x)).map('${:,.2f}'.format)
#df_topclientsummary["TotalProfit"] = df_topclientsummary["TotalProfit"].apply(lambda x: Decimal(x)).map('${:,.2f}'.format)
#df_topclientsummary

def dollartomillions(amount):
    millions = (amount) / 100000
    return f"${millions:,.2f}M"
    

# Define a function that converts a dollar amount to millions.

# Apply the currency_format_millions function to only the money columns. 
df_topclientsummary["TotalRevenueInMillions"]=df_topclientsummary["TotalRevenue"].apply(lambda x: dollartomillions(x))
df_topclientsummary["TotalProfitInMillions"]=df_topclientsummary["TotalProfit"].apply(lambda x: dollartomillions(x))
df_topclientsummary["TotalRevenue"] = df_topclientsummary["TotalRevenue"].apply(lambda x: Decimal(x)).map('${:,.2f}'.format)
df_topclientsummary["TotalProfit"] = df_topclientsummary["TotalProfit"].apply(lambda x: Decimal(x)).map('${:,.2f}'.format)



# Rename the columns to reflect the change in the money format. 
df_topclientsummary.rename(columns = {'client_id': 'Client ID','TotalUnitsPurchased': 'Total Units Purchased','TotalShippingPrice':'Total Shipping Price','TotalRevenue':'Total Revenue','TotalProfit': 'Total Profit','TotalRevenueInMillions': 'Total Revenue (millions)','TotalProfitInMillions': 'Total Profit (millions)'},inplace=True)
df_topclientsummary

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_topclientsummary["Total Profit (millions)"] = df_topclientsummary["Total Profit (millions)"].replace({'\$':'',',':'','M':''},regex=True).astype(float)
#df_topclientsummary
new_df = df_topclientsummary.sort_values("Total Profit (millions)",ascending=False)
new_df["Total Profit (millions)"]= new_df["Total Profit (millions)"].map('${:,.2f}M'.format)
new_df

```
